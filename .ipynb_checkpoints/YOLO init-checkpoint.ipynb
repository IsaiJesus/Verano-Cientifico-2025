# Preparación del dataset

```python
import os
import shutil
import random
from pathlib import Path

def split_dataset(source_dir, dest_dir, train_ratio=0.7, val_ratio=0.2, test_ratio=0.1):
    """
    Separa un dataset organizado por clases en carpetas train/val/test
    
    Args:
        source_dir: Directorio fuente con subcarpetas por clase
        dest_dir: Directorio destino para el dataset separado
        train_ratio: Proporción para entrenamiento (default: 0.7)
        val_ratio: Proporción para validación (default: 0.2)
        test_ratio: Proporción para prueba (default: 0.1)
    """
    
    # Verificar que las proporciones sumen 1
    if abs(train_ratio + val_ratio + test_ratio - 1.0) > 0.001:
        raise ValueError("Las proporciones deben sumar 1.0")
    
    source_path = Path(source_dir)
    dest_path = Path(dest_dir)
    
    # Crear directorios destino
    for split in ['train', 'val', 'test']:
        (dest_path / split).mkdir(parents=True, exist_ok=True)
    
    # Procesar cada clase
    for class_dir in source_path.iterdir():
        if not class_dir.is_dir():
            continue
            
        class_name = class_dir.name
        print(f"Procesando clase: {class_name}")
        
        # Crear subdirectorios para cada clase en train/val/test
        for split in ['train', 'val', 'test']:
            (dest_path / split / class_name).mkdir(parents=True, exist_ok=True)
        
        # Obtener todas las imágenes de la clase
        image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp'}
        images = [f for f in class_dir.iterdir() 
                 if f.suffix.lower() in image_extensions]
        
        if not images:
            print(f"  Advertencia: No se encontraron imágenes en {class_dir}")
            continue
        
        # Mezclar las imágenes aleatoriamente
        random.shuffle(images)
        
        # Calcular índices de separación
        total = len(images)
        train_end = int(total * train_ratio)
        val_end = train_end + int(total * val_ratio)
        
        # Separar imágenes
        train_images = images[:train_end]
        val_images = images[train_end:val_end]
        test_images = images[val_end:]
        
        print(f"  Train: {len(train_images)}, Val: {len(val_images)}, Test: {len(test_images)}")
        
        # Copiar imágenes a sus respectivos directorios
        for img_list, split in [(train_images, 'train'), 
                               (val_images, 'val'), 
                               (test_images, 'test')]:
            for img_path in img_list:
                dest_img_path = dest_path / split / class_name / img_path.name
                shutil.copy2(img_path, dest_img_path)

# Uso del script
if __name__ == "__main__":
    # Configurar rutas
    source_directory = "./data/Augmented Dataset"  # Cambia esta ruta
    destination_directory = "./data/dataset"  # Cambia esta ruta
    
    # Establecer semilla para reproducibilidad
    random.seed(42)
    
    # Separar dataset
    split_dataset(
        source_dir=source_directory,
        dest_dir=destination_directory,
        train_ratio=0.7,  # 70% para entrenamiento
        val_ratio=0.2,    # 20% para validación
        test_ratio=0.1    # 10% para prueba
    )
    
    print("Dataset separado exitosamente!")
```

```python
import os
import yaml
from pathlib import Path

def create_yolo_config(dataset_dir, config_name="dataset_config.yaml"):
    """
    Crea un archivo de configuración YAML para YOLOv8
    
    Args:
        dataset_dir: Directorio del dataset separado
        config_name: Nombre del archivo de configuración
    """
    
    dataset_path = Path(dataset_dir)
    
    # Obtener las clases desde el directorio train
    train_dir = dataset_path / "train"
    if not train_dir.exists():
        raise ValueError(f"No se encontró el directorio train en {dataset_path}")
    
    # Extraer nombres de clases
    classes = [d.name for d in train_dir.iterdir() if d.is_dir()]
    classes.sort()  # Ordenar para consistencia
    
    # Crear configuración
    config = {
        'path': str(dataset_path.absolute()),  # Ruta absoluta al dataset
        'train': 'train',  # Directorio de entrenamiento (relativo a 'path')
        'val': 'val',      # Directorio de validación (relativo a 'path')
        'test': 'test',    # Directorio de prueba (relativo a 'path')
        'nc': len(classes),  # Número de clases
        'names': classes     # Lista de nombres de clases
    }
    
    # Guardar configuración
    config_path = dataset_path / config_name
    with open(config_path, 'w', encoding='utf-8') as f:
        yaml.dump(config, f, default_flow_style=False, allow_unicode=True)
    
    print(f"Archivo de configuración creado: {config_path}")
    print(f"Clases detectadas ({len(classes)}): {classes}")
    
    return str(config_path)

# Ejemplo de uso
if __name__ == "__main__":
    dataset_directory = "./data/dataset"  # Cambia esta ruta
    config_file = create_yolo_config(dataset_directory)
    
    # Mostrar contenido del archivo generado
    with open(config_file, 'r', encoding='utf-8') as f:
        print("\nContenido del archivo de configuración:")
        print(f.read())
```

```python
from ultralytics import YOLO
import torch

def train_yolo_classifier(data_path, model_size='n', epochs=50, imgsz=800, batch_size=16):
    """
    Entrena un modelo YOLOv8 para clasificación
    
    Args:
        data_path: Ruta al directorio del dataset (NO al archivo YAML)
        model_size: Tamaño del modelo ('n', 's', 'm', 'l', 'x')
        epochs: Número de épocas de entrenamiento
        imgsz: Tamaño de imagen para entrenamiento
        batch_size: Tamaño del lote
    """
    
    # Verificar si CUDA está disponible
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Usando device: {device}")
    
    # Cargar modelo preentrenado para clasificación
    model_name = f'yolov8{model_size}-cls.pt'
    model = YOLO(model_name)
    
    print(f"Modelo cargado: {model_name}")
    print(f"Directorio del dataset: {data_path}")
    
    # Entrenar el modelo - NOTA: usar directorio, no archivo YAML
    results = model.train(
        data=data_path,              # Directorio del dataset, NO archivo YAML
        epochs=epochs,
        imgsz=imgsz,
        batch=batch_size,
        device=device,
        project='runs/classify',
        name='train',
        save=True,
        save_period=10,
        cache=True,
        workers=8,
        patience=50,
        optimizer='AdamW',
        lr0=0.001,
        weight_decay=0.0005,
        warmup_epochs=3,
        warmup_momentum=0.8,
        cos_lr=True,
        close_mosaic=10,
        amp=True,
        fraction=1.0,
        profile=False,
        freeze=None,
        multi_scale=True,
        overlap_mask=True,
        mask_ratio=4,
        dropout=0.0,
        val=True,
        plots=True,
        verbose=True
    )
    
    print("Entrenamiento completado!")
    print(f"Mejor modelo guardado en: {model.trainer.best}")
    
    return results


if __name__ == "__main__":
    # CAMBIO PRINCIPAL: usar directorio del dataset
    dataset_directory = r"C:\Users\isaij\Documents\Projects\Verano Científico\Proyecto\data\dataset"
    
    # Entrenar modelo
    print("Iniciando entrenamiento...")
    training_results = train_yolo_classifier(
        data_path=dataset_directory,  # Directorio, no YAML
        model_size='n',
        epochs=50,
        imgsz=800,
        batch_size=16
    )
    
    # El modelo entrenado se guardará automáticamente
    # Ubicación típica: runs/classify/train/weights/best.pt
    
    # Evaluar modelo (opcional)
    # best_model_path = "runs/classify/train/weights/best.pt"
    # eval_results = evaluate_model(best_model_path, dataset_directory)
    
    # Hacer predicción en imagen individual (opcional)
    # test_image = "path/to/test/image.jpg"
    # prediction = predict_single_image(best_model_path, test_image)
```

```python
def evaluate_model(model_path, test_data_path):
    """
    Evalúa el modelo entrenado
    
    Args:
        model_path: Ruta al modelo entrenado
        test_data_path: Ruta al directorio de datos de prueba
    """
    
    # Cargar modelo entrenado
    model = YOLO(model_path)
    
    # Evaluar en datos de prueba
    results = model.val(data=test_data_path, split='test')
    
    print("Evaluación completada!")
    print(f"Accuracy: {results.top1}")
    print(f"Top-5 Accuracy: {results.top5}")
    
    return results
```

```python
def predict_single_image(model_path, image_path):
    """
    Hace predicción en una imagen individual
    
    Args:
        model_path: Ruta al modelo entrenado
        image_path: Ruta a la imagen
    """
    
    # Cargar modelo
    model = YOLO(model_path)
    
    # Hacer predicción
    results = model(image_path)
    
    # Mostrar resultados
    for result in results:
        probs = result.probs
        print(f"Clase predicha: {result.names[probs.top1]}")
        print(f"Confianza: {probs.top1conf:.4f}")
        print(f"Top 5 predicciones:")
        for i, (class_id, conf) in enumerate(zip(probs.top5, probs.top5conf)):
            print(f"  {i+1}. {result.names[class_id]}: {conf:.4f}")
    
    return results
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ultralytics import YOLO
from sklearn.metrics import confusion_matrix, classification_report
import cv2
from pathlib import Path
import pandas as pd

def create_confusion_matrix(model_path, test_data_path, save_path="confusion_matrix.png"):
    """
    Crea y visualiza la matriz de confusión
    
    Args:
        model_path: Ruta al modelo entrenado
        test_data_path: Ruta al directorio de datos de prueba
        save_path: Ruta donde guardar la imagen de la matriz
    """
    
    # Cargar modelo
    model = YOLO(model_path)
    
    # Obtener todas las imágenes de test y sus etiquetas verdaderas
    test_path = Path(test_data_path)
    
    y_true = []
    y_pred = []
    class_names = []
    
    # Recorrer todas las carpetas de clases en el directorio de test
    for class_folder in test_path.iterdir():
        if class_folder.is_dir():
            class_name = class_folder.name
            if class_name not in class_names:
                class_names.append(class_name)
            
            # Procesar cada imagen en la carpeta de la clase
            for img_file in class_folder.glob("*.jpg"):
                try:
                    # Hacer predicción
                    results = model(str(img_file))
                    
                    # Obtener la clase predicha
                    for result in results:
                        predicted_class = result.names[result.probs.top1]
                        
                        y_true.append(class_name)
                        y_pred.append(predicted_class)
                        
                except Exception as e:
                    print(f"Error procesando {img_file}: {e}")
                    continue
    
    # Crear matriz de confusión
    cm = confusion_matrix(y_true, y_pred, labels=class_names)
    
    # Visualizar matriz de confusión
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Matriz de Confusión')
    plt.xlabel('Predicción')
    plt.ylabel('Valor Real')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    
    # Imprimir reporte de clasificación
    print("\nReporte de Clasificación:")
    print(classification_report(y_true, y_pred, target_names=class_names))
    
    return cm, y_true, y_pred

def plot_training_metrics(runs_path="runs/classify/train"):
    """
    Visualiza las métricas de entrenamiento desde los resultados de YOLO
    
    Args:
        runs_path: Ruta a los resultados del entrenamiento
    """
    
    results_csv = Path(runs_path) / "results.csv"
    
    if not results_csv.exists():
        print(f"No se encontró el archivo de resultados en {results_csv}")
        return
    
    # Leer datos de entrenamiento
    df = pd.read_csv(results_csv)
    df.columns = df.columns.str.strip()  # Limpiar espacios en nombres de columnas
    
    # Crear subplots
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Plot 1: Loss de entrenamiento y validación
    if 'train/loss' in df.columns and 'val/loss' in df.columns:
        axes[0, 0].plot(df['epoch'], df['train/loss'], label='Train Loss', color='blue')
        axes[0, 0].plot(df['epoch'], df['val/loss'], label='Val Loss', color='red')
        axes[0, 0].set_title('Pérdida durante el entrenamiento')
        axes[0, 0].set_xlabel('Época')
        axes[0, 0].set_ylabel('Pérdida')
        axes[0, 0].legend()
        axes[0, 0].grid(True)
    
    # Plot 2: Accuracy
    if 'metrics/accuracy_top1' in df.columns:
        axes[0, 1].plot(df['epoch'], df['metrics/accuracy_top1'], 
                       label='Top-1 Accuracy', color='green')
        if 'metrics/accuracy_top5' in df.columns:
            axes[0, 1].plot(df['epoch'], df['metrics/accuracy_top5'], 
                           label='Top-5 Accuracy', color='orange')
        axes[0, 1].set_title('Precisión durante el entrenamiento')
        axes[0, 1].set_xlabel('Época')
        axes[0, 1].set_ylabel('Precisión')
        axes[0, 1].legend()
        axes[0, 1].grid(True)
    
    # Plot 3: Learning Rate
    if 'lr/pg0' in df.columns:
        axes[1, 0].plot(df['epoch'], df['lr/pg0'], color='purple')
        axes[1, 0].set_title('Tasa de Aprendizaje')
        axes[1, 0].set_xlabel('Época')
        axes[1, 0].set_ylabel('Learning Rate')
        axes[1, 0].grid(True)
    
    # Plot 4: Métricas adicionales si están disponibles
    available_metrics = [col for col in df.columns if 'metrics/' in col and 'accuracy' not in col]
    if available_metrics:
        for metric in available_metrics[:3]:  # Mostrar hasta 3 métricas adicionales
            axes[1, 1].plot(df['epoch'], df[metric], label=metric.replace('metrics/', ''))
        axes[1, 1].set_title('Métricas Adicionales')
        axes[1, 1].set_xlabel('Época')
        axes[1, 1].set_ylabel('Valor')
        axes[1, 1].legend()
        axes[1, 1].grid(True)
    
    plt.tight_layout()
    plt.savefig('training_metrics.png', dpi=300, bbox_inches='tight')
    plt.show()

def visualize_predictions(model_path, test_images_path, num_images=6, save_path="predictions_visualization.png"):
    """
    Visualiza predicciones en un conjunto de imágenes de prueba
    
    Args:
        model_path: Ruta al modelo entrenado
        test_images_path: Ruta a las imágenes de prueba
        num_images: Número de imágenes a mostrar
        save_path: Ruta donde guardar la visualización
    """
    
    # Cargar modelo
    model = YOLO(model_path)
    
    # Obtener lista de imágenes de prueba
    test_path = Path(test_images_path)
    image_files = []
    
    # Buscar imágenes en todas las subcarpetas
    for img_format in ['*.jpg', '*.jpeg', '*.png']:
        image_files.extend(list(test_path.rglob(img_format)))
    
    # Seleccionar imágenes aleatorias
    selected_images = np.random.choice(image_files, min(num_images, len(image_files)), replace=False)
    
    # Crear subplot
    cols = 3
    rows = (num_images + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols, figsize=(15, 5*rows))
    
    if rows == 1:
        axes = axes.reshape(1, -1)
    
    for i, img_path in enumerate(selected_images):
        row = i // cols
        col = i % cols
        
        # Cargar y mostrar imagen
        img = cv2.imread(str(img_path))
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Hacer predicción
        results = model(str(img_path))
        
        # Obtener predicción
        for result in results:
            predicted_class = result.names[result.probs.top1]
            confidence = result.probs.top1conf.item()
            
            # Obtener clase verdadera del nombre de la carpeta
            true_class = img_path.parent.name
            
            # Mostrar imagen
            axes[row, col].imshow(img_rgb)
            axes[row, col].set_title(f'Real: {true_class}\nPredicción: {predicted_class}\nConfianza: {confidence:.3f}')
            axes[row, col].axis('off')
    
    # Ocultar axes vacíos
    for i in range(len(selected_images), rows * cols):
        row = i // cols
        col = i % cols
        axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()

def plot_class_distribution(test_data_path, save_path="class_distribution.png"):
    """
    Visualiza la distribución de clases en el dataset de prueba
    
    Args:
        test_data_path: Ruta al directorio de datos de prueba
        save_path: Ruta donde guardar la gráfica
    """
    
    test_path = Path(test_data_path)
    class_counts = {}
    
    # Contar imágenes por clase
    for class_folder in test_path.iterdir():
        if class_folder.is_dir():
            class_name = class_folder.name
            img_count = len(list(class_folder.glob("*.jpg"))) + len(list(class_folder.glob("*.png")))
            class_counts[class_name] = img_count
    
    # Crear gráfica de barras
    plt.figure(figsize=(12, 6))
    classes = list(class_counts.keys())
    counts = list(class_counts.values())
    
    bars = plt.bar(classes, counts, color='skyblue', edgecolor='navy', alpha=0.7)
    plt.title('Distribución de Clases en Dataset de Prueba')
    plt.xlabel('Clases')
    plt.ylabel('Número de Imágenes')
    plt.xticks(rotation=45)
    
    # Añadir valores en las barras
    for bar, count in zip(bars, counts):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
                str(count), ha='center', va='bottom')
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    
    return class_counts

def plot_confidence_distribution(model_path, test_data_path, save_path="confidence_distribution.png"):
    """
    Visualiza la distribución de confianza de las predicciones
    
    Args:
        model_path: Ruta al modelo entrenado
        test_data_path: Ruta al directorio de datos de prueba
        save_path: Ruta donde guardar la gráfica
    """
    
    # Cargar modelo
    model = YOLO(model_path)
    
    # Obtener todas las confianzas
    confidences = []
    correct_confidences = []
    incorrect_confidences = []
    
    test_path = Path(test_data_path)
    
    for class_folder in test_path.iterdir():
        if class_folder.is_dir():
            class_name = class_folder.name
            
            for img_file in class_folder.glob("*.jpg"):
                try:
                    results = model(str(img_file))
                    
                    for result in results:
                        predicted_class = result.names[result.probs.top1]
                        confidence = result.probs.top1conf.item()
                        
                        confidences.append(confidence)
                        
                        if predicted_class == class_name:
                            correct_confidences.append(confidence)
                        else:
                            incorrect_confidences.append(confidence)
                            
                except Exception as e:
                    continue
    
    # Crear histograma
    plt.figure(figsize=(12, 8))
    
    # Subplot 1: Distribución general
    plt.subplot(2, 1, 1)
    plt.hist(confidences, bins=30, alpha=0.7, color='blue', edgecolor='black')
    plt.title('Distribución de Confianza - Todas las Predicciones')
    plt.xlabel('Confianza')
    plt.ylabel('Frecuencia')
    plt.grid(True, alpha=0.3)
    
    # Subplot 2: Correctas vs Incorrectas
    plt.subplot(2, 1, 2)
    plt.hist(correct_confidences, bins=20, alpha=0.7, color='green', 
             label=f'Correctas ({len(correct_confidences)})', edgecolor='black')
    plt.hist(incorrect_confidences, bins=20, alpha=0.7, color='red', 
             label=f'Incorrectas ({len(incorrect_confidences)})', edgecolor='black')
    plt.title('Distribución de Confianza - Correctas vs Incorrectas')
    plt.xlabel('Confianza')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    
    return confidences, correct_confidences, incorrect_confidences

def create_comprehensive_report(model_path, test_data_path):
    """
    Genera un reporte visual completo del modelo
    
    Args:
        model_path: Ruta al modelo entrenado
        test_data_path: Ruta al directorio de datos de prueba
    """
    
    print("="*60)
    print("GENERANDO REPORTE VISUAL COMPLETO")
    print("="*60)
    
    print("\n1. Creando matriz de confusión...")
    cm, y_true, y_pred = create_confusion_matrix(model_path, test_data_path)
    
    print("\n2. Visualizando métricas de entrenamiento...")
    plot_training_metrics()
    
    print("\n3. Mostrando predicciones de ejemplo...")
    visualize_predictions(model_path, test_data_path)
    
    print("\n4. Analizando distribución de clases...")
    class_counts = plot_class_distribution(test_data_path)
    
    print("\n5. Evaluando distribución de confianza...")
    confidences, correct_conf, incorrect_conf = plot_confidence_distribution(model_path, test_data_path)
    
    # Estadísticas finales
    print("\n" + "="*60)
    print("ESTADÍSTICAS FINALES")
    print("="*60)
    print(f"Total de imágenes evaluadas: {len(y_true)}")
    print(f"Precisión general: {sum(np.array(y_true) == np.array(y_pred)) / len(y_true):.4f}")
    print(f"Confianza promedio: {np.mean(confidences):.4f}")
    print(f"Confianza promedio (correctas): {np.mean(correct_conf):.4f}")
    print(f"Confianza promedio (incorrectas): {np.mean(incorrect_conf):.4f}")
    
    print("\nDistribución de clases:")
    for class_name, count in class_counts.items():
        print(f"  {class_name}: {count} imágenes")

# Ejemplo de uso
if __name__ == "__main__":
    # Configurar rutas
    model_path = "runs/classify/train/weights/best.pt"
    test_data_path = "data/dataset/test"
    
    # Generar todas las visualizaciones
    # create_comprehensive_report(model_path, test_data_path)
    
    # O usar funciones individuales:
    create_confusion_matrix(model_path, test_data_path)
    plot_training_metrics()
    # visualize_predictions(model_path, test_data_path)
    # plot_class_distribution(test_data_path)
    # plot_confidence_distribution(model_path, test_data_path)
```
